```python
GitLinkXLS='https://docs.google.com/spreadsheets/d/e/2PACX-1vQQ_ePBaN0RuAw0aHdA4tNwScjDe6v6m2O342Hv9bekJyJADPmj_1Qg0VJwapzC4bZosu15CRhePpca/pub?gid=0&single=true&output=csv'

Education=pd.read_csv(GitLinkXLS).drop(columns=['EIRank','ExpectedYears','MeanYears','HDI','Continent'])
Education
```

```python
Education.Country.tolist()
```

```python
from urllib.request import urlopen

GitLinkXLS='https://github.com/tygris/PUBPOL_GovAnalytics/raw/master/tax.pkl'
Tax=pd.read_pickle(urlopen(GitLinkXLS),compression=None)

country_fix = {'Czechia': 'Czech Republic',
               'Korea, South': 'South Korea',
               'Russia': 'Russian Federation',
               'Brunei': 'Brunei Darussalam',
               'Bahamas, The': 'Bahamas',
               'West Bank': 'Palestine',
               'Micronesia, Federated States of': 'Micronesia',
               'Vietnam': 'Viet Nam',
               'Eswatini': 'Swaziland',
               'Congo, Republic of the': 'Congo',
               'Congo, Democratic Republic of the': 'Democratic Republic of the Congo',
               'Syria': 'Syrian Arab Republic',
               'Burma': 'Myanmar',
               'Gambia, The': 'Gambia'}

# use the dictionary
Tax.Country.replace(country_fix,inplace=True)

Tax.head()
```

```python
Tax.Country.tolist()
```

```python
GitLinkXLS='https://raw.githubusercontent.com/abunnell-source/GroupData/master/DFnew.csv'
Religion=pd.read_csv(GitLinkXLS).loc[:,['Country','ReligionImportant']]

# dictionary to change 'Country' names
country_fix2 = {'Vietnam': 'Viet Nam',
 'The Netherlands': 'Netherlands',
 'Russia': 'Russian Federation',
 'North Macedonia': 'Macedonia',
 'Syria': 'Syrian Arab Republic',
 'State of Palestine': 'Palestine',
 'Republic of the Congo': 'Congo'}
# use the dictionary

Religion.Country.replace(country_fix2,inplace=True)

Religion.head()
```

```python
Religion.Country.tolist()
```

```python
len(Education), len(Tax), len(Religion)
```

```python
Data=Education.merge(Tax,how='inner').merge(Religion,how='inner')
```

```python
Data.columns
```

```python
Data=Data.dropna()
Data.reset_index(drop=True,inplace=True)
Data.head()
```

```python
Data.to_csv('Data.csv', index=False)
```
